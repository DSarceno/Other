```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

url_base = 'https://www.football-data.co.uk/'
url = url_base + 'downloadm.php'

try:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
except Exception as e:
    print(e)


useful_links = []
pattern = 'all-euro-data'
links = soup.find_all('a', href=True)
for link in links:
    l = link['href']
    match1 = re.search(pattern, l)
    if match1:
        print(f'Link: {url_base + l}')
        useful_links.append(url_base + l)

dfs = []
for link in useful_links:
    all_sheets = pd.read_excel(link, sheet_name=None)
    for _, df in all_sheets.items():
        dfs.append(df)

df = pd.concat(dfs, ignore_index=True)

df.to_excel('all-leagues-data.xlsx', index=False)
```

```python
dfuseful = df[['Div', 'Date', 'Time', 'HomeTeam', 'AwayTeam', 'FTHG', 'FTAG', 'FTR',
       'HTHG', 'HTAG', 'HTR', 'Referee', 'HS', 'AS', 'HST', 'AST', 'HF', 'AF',
       'HC', 'AC', 'HY', 'AY', 'HR', 'AR']]
```

```python
dfuseful = dfuseful.dropna(how='all')
```

```python
dfuseful.info()
```

```python
dfuseful.to_excel('all-leagues-data.xlsx', index=False)
```

```python
df2 = dfuseful
```

```python
str(df2['Date'].dt.month + 1)
```

```python
df2['Season'] = np.where(
    df2['Date'].dt.month > 7,
    df2['Date'].dt.year.astype(str) + "-" + (df2['Date'].dt.year + 1).astype(str),
    (df2['Date'].dt.year - 1).astype(str) + "-" + df2['Date'].dt.year.astype(str)
)
```

```python
def calculate_points(row):
    if row['FTR'] == 'H':
        return 3, 0
    elif row['FTR'] == 'A':
        return 0, 3
    else:
        return 1, 1
```

```python
df2['HomeTeam_Points'] = 0
df2['AwayTeam_Points'] = 0
df2['HomeTeam_Games'] = 0
df2['AwayTeam_Games'] = 0
```

```python
for season in df2['Season'].unique():
    # filtramos primero por temporada
    sdata = df2[df2['Season'] == season].sort_values(by='Date')

    # diccionarios para llevar el control de puntos y partidos jugados
    points = {}
    games_played = {}

    # iteramos por la data
    for index, row in sdata.iterrows():
        home_team = row['HomeTeam']
        away_team = row['AwayTeam']


        # inicializamos los puntos y partidos en caso de que no lo esten
        if home_team not in points:
            points[home_team] = 0
            games_played[home_team] = 0
        if away_team not in points:
            points[away_team] = 0
            games_played[away_team] = 0
    
        
        # calculamos los putnos ganados este partido
        home_points, away_points = calculate_points(row)

        
        # actualizamos los puntos y partidos jugados
        points[home_team] += home_points
        games_played[home_team] += 1
        points[away_team] += away_points
        games_played[away_team] += 1



        # asignar los puntos luego del partido
        df.at[index, 'HomeTeam_Points'] = points[home_team]
        df.at[index, 'AwayTeam_Points'] = points[away_team]
        df.at[index, 'HomeTeam_Games'] = games_played[home_team]
        df.at[index, 'AwayTeam_Games'] = games_played[away_team]
```

```python
df2.to_excel('new-all-football-data.xlsx', index=False)
```
