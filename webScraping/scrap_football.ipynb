```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
url_base = 'https://www.football-data.co.uk/'
url = url_base + 'downloadm.php'
try:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, 'html.parser')
except Exception as e:
    print(e)
```

```python
useful_links = []
pattern = 'all-euro-data'
links = soup.find_all('a', href=True)
for link in links:
    l = link['href']
    match1 = re.search(pattern, l)
    if match1:
        print(f'Link: {url_base + l}')
        useful_links.append(url_base + l)
```

```python
df = pd.read_excel(useful_links[0], sheet_name='SP1')
```
